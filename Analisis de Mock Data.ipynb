**Medidas de tendencia central**

```python
#Calcular la media de mi conjunto
media_age = df['age'].mean()
print(f'La media de la edad es: {media_age}')
```

```python
#Calcular la mediana de mi conjunto
mediana_age = df['age'].median()
print(f'La mediana de la edad es: {mediana_age}')
```

```python
#Calcular la moda de mi conjunto
moda_age = df['age'].mode()
print(f'La moda de la edad es: {moda_age}')
```

Calcular media, mediana y moda de service_call_duration

```python
#Calcular la media de mi conjunto
media_service_call_duration = df['service_call_duration'].mean()
print(f'La media de la duracion de la llamada de servicio es: {media_service_call_duration}')
```

```python
#Calcular la mediana de mi conjunto
mediana_service_call_duration = df['service_call_duration'].median()
print(f'La mediana de la duracion de la llamada de servicio es: {mediana_service_call_duration}')
```

```python
#Calcular la moda de mi conjunto
moda_service_call_duration = df['service_call_duration'].mode()
print(f'La moda de la duracion de la llamada de servicio es: {moda_service_call_duration}')
```

**Medidas de dispersi√≥n**

```python
#Calcular el rango para age
rango_age = df['age'].max() - df['age'].min()
print(f'El rango de la edad es: {rango_age}')
```

```python
#Calcular la varianza de age
varianza_age = df['age'].var()
print(f'La varianza de la edad es: {varianza_age}')
```

```python
#Desviaci√≥n estandar
desv_est_age =df['age'].std()
desv_est_raiz_age = np.sqrt(varianza_age)
print(f'La desviaci√≥n estandar de la edad es: {desv_est_age}')
print(f'La desviaci√≥n estandar con la raiz de la edad es: {desv_est_raiz_age}')
```

```python
#Coeficiente de variaci√≥n
cv_age =(desv_est_age/media_age)*100
print(f'El coeficiente de variacion de la edad es: {cv_age:.2f}%')
```

```python
#Identificar las frecuencias y datos unicos de cada elemento de la columna o campo
tabla_frec = df['age'].value_counts().sort_index()
plt.figure(figsize=[10,7])
plt.bar(tabla_frec.index, tabla_frec.values, color='red', edgecolor = 'black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribucion de las frecuencias de las edades')
plt.show()
print(tabla_frec)
```
